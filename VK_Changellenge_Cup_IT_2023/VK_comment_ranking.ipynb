### **Лемматизация текстов**

---

**Расширим список стоп-слов:**

```python
english_stopwords = stopwords.words('english')
english_stopwords.extend(['x', 'f', 'i', 'the', 'http', 'it', 'com', 'could', 'www', 'would', 'quot']) 
```

---

**Создадим функцию для лемматизации текстов:**

```python
def lem_text(text):
    
    final = []
    for t in notebook.tqdm(text):
        t = " ".join(re.sub(r'[^a-zA-Z]', ' ', t).split()) # оставляем только буквы, избавляемся от лишних символов
        token_list = word_tokenize(t) # токенизируем
        
        stop_text = []
        for word in token_list: 
            if word.lower() not in english_stopwords: # удаляем стоп-слова
                stop_text.append(word.lower())
        
        lemmatizer = WordNetLemmatizer()
        lemm_list = []
        for w in stop_text:
            lem_t = lemmatizer.lemmatize(w) # выполняем лемматизацию каждого слова
            lemm_list.append(lem_t)
        lemm_list = " ".join(lemm_list)
        final.append(lemm_list)
        
    return final
```

---

**Преобразуем признаки в тренировочной и тестовых выборках:**

### **Создание скоринга**

```python
from sklearn.metrics import make_scorer
```

```python
def ndcg(target, predict):
    true_relevance = np.asarray([target])
    scores = np.asarray([predict])
    return ndcg_score(true_relevance, scores)
```

```python
custom_score = make_scorer(ndcg)
```

### **LogisticRegression**

---

---

```python
def log_reg(x_train, y_train):
    pipe_lr = Pipeline([('tfidf', TfidfVectorizer()),
                ('clf', LogisticRegression(random_state=RS))])
 
    scores = cross_val_score(pipe_lr, x_train, y_train,
                                     scoring = custom_score, cv=3, n_jobs=-1).mean()
 
    model_lr = pipe_lr.fit(x_train, y_train)
    print("ndcg =", scores)
    return model_lr
```

```python
def tree(x_train, y_train):
    model = Pipeline([('tfidf', TfidfVectorizer()),
                 ('clf', DecisionTreeClassifier(random_state=RS))])
    
    parametrs = {'clf__max_depth': range (1, 15, 3), 
                 'clf__min_samples_split': range (2, 9, 2)}
    
    model_tree = GridSearchCV(model, parametrs, cv=3, scoring = custom_score, n_jobs=-1)
    model_tree.fit(x_train, y_train)
    return model_tree
```

**Выполним шаги:**
- получим промежуточную структуру файла, по аналогии с пунктом `2.2`
- ранжируем предсказания для каждых 5 комментариев в порядке возрастания скора, одновременно с этим вернем исходную структуру файла в формате `.jsonl`

---

**На этом этапе мы снова получим промежуточную структуру файлов, которая была в начале исследования в пункте 2.2**

```python
all_back = pd.DataFrame({'text':final_df_test['text'][0:final_df_test.shape[0]:5].values})

def to_many(dataframe, rank, number_fo_ranks):
    comment_rank = final_df_test[rank:dataframe.shape[0]*number_fo_ranks:number_fo_ranks]
    comment_rank.reset_index(drop=True, inplace=True)
    comment_rank.drop('text', axis=1, inplace=True)
    comment_rank = comment_rank.rename(columns={'comment':'comment_{}'.format(rank), 'score':'score_{}'.format(rank)})
    return comment_rank

for i in range(0, 5):
    all_back = all_back.join(to_many(final_df_test, i, 5))   
```

```python
all_back.head(5)
```

---

---

**Код в последующих ячейках ранжирует каждые 5 комментариев, относящиеся к уникальному тексту, в зависимости от `предсказанного моделью score` и возвращает структуру файла jsonl.**

*Код создает итоговый файл с ранжированными комментариями по score и самой заполненной колонкой score.*

---

```python
all_back['comments'] = 'null'

for index, row in all_back.iterrows():
    comments = []
    for i in range(0, 5):
        comments.append({'text':str(row['comment_{}'.format(i)]), 'score': int(row['score_{}'.format(i)])})
    comments = sorted(comments, key=lambda x: x.get('score'), reverse=False)
    comments_score_by_index = []
    for i in range(0, 5):
        comment = comments[i]
        comment['score'] = i
        comments_score_by_index.append(comment)
    all_back.at[index, 'comments'] = json.loads(json.dumps([json.loads(json.dumps(comment)) for comment in comments_score_by_index]))
```

```python
all_back_almost_ready = all_back[['text', 'comments']]
```

```python
out = all_back_almost_ready.to_json(orient='records', lines=True)

with open('ranking_test_final.jsonl', 'w') as f: # указать папку выгрузки
    f.write(out)    
```
